```python
# Python program to read image using OpenCV

# importing OpenCV(cv2) module
import cv2

img = cv2.imread('tile.png')
```

```python
from matplotlib import pyplot as plt

plt.imshow(img)

plt.show()
```

```python
import numpy as np
rgb = np.array(img)
```

```python
print(rgb)
shape=np.shape(rgb)
print()
rows = shape[0];
cols = shape[1];
```

```python
ans = []
for i in range(rows):
  for j in range(cols):
    if(rgb[i][j][1]>252):
      ans.append([i,j])
    #print(rgb[i][j])
```
